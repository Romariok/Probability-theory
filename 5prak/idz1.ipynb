```python
max_value = np.max(data)
min_value = np.min(data)
spread = max_value - min_value
```

```python
start = 15
leg = 16
data_interval = []
for i in range(9):
   bound1 = start + i * leg
   bound2 = start + (i+1) * leg
   count_interval = np.count_nonzero((data >= bound1) & (data <= bound2))
   print(f"Начало: {bound1}, Конец: {bound2}\t Середина: {(bound2-bound1)/2+bound1} \t Частота интервалов:{count_interval}")
   print(f"Относительная частота: {count_interval/100} \t Плотность относительнйо частоты: {(count_interval/100)/16}")
   print("\n")
   data_interval.append([bound1, bound2,(bound2-bound1)/2+bound1, count_interval,count_interval/100, (count_interval/100)/16])

data_interval
```

```python
c1=0
c3=0
for interval in data_interval:
   c1tmp = interval[3]* interval[2]
   c3tmp = interval[3]* interval[2]*interval[2]
   c1+=c1tmp
   c3+=c3tmp
   print(f"nx = {c1tmp}\tx^2 = {interval[2]**2}\tnx^2= {c3tmp}")


print(c1)
print(c3)
mean = c1/100
dispersion = c3/100 - mean*mean
print(f"Mean: {mean}")
print(f"Dispersion: {dispersion}")
print(f"SKO:{np.sqrt(c3/100 - mean*mean)}")
n_sko = np.sqrt(len(data)*dispersion/len(data))
print(f"nice dispersion: {len(data)*dispersion/len(data)}")
print(f"nice SKO: {np.sqrt(len(data)*dispersion/len(data))}")
```

```python

# Извлечение значений для осей x и y
x = [interval[2] for interval in data_interval]
y = [interval[3] for interval in data_interval]

# Построение точек
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Настройка сетки
plt.grid(True)
plt.xticks(x)
plt.yticks(y)

# Установка подписей осей и заголовка
plt.xlabel('Значение x')
plt.ylabel('Значение y')
plt.title('Полигон с сеткой')

# Отображение графика
plt.show()
```

```python

# Извлечение значений для осей x, y и высоты столбцов
x = [interval[0] for interval in data_interval]
y = [interval[1] for interval in data_interval]
heights = [interval[5] for interval in data_interval]

# Построение гистограммы
plt.bar(x, heights, width=16, align='edge')

plt.grid(True)
plt.xticks(x)
plt.yticks(heights)

# Установка подписей осей и заголовка
plt.xlabel('Начало интервала')
plt.ylabel('Высота столбцов')
plt.title('Гистограмма')

# Отображение графика
plt.show()
```

```python

# Извлечение значений, для которых нужно построить ЭФР
x = [interval[2] for interval in data_interval]

# Сортировка значений
x_sorted = np.sort(x)

# Вычисление вероятностей
n = len(x_sorted)
probabilities = np.arange(1, n+1) / n

plt.grid(True)
plt.xticks(x)
plt.yticks(probabilities)

# Построение графика
plt.step(x_sorted, probabilities, where='post')

# Установка подписей осей и заголовка
plt.xlabel('Значение x')
plt.ylabel('Эмпирическая функция распределения')
plt.title('Эмпирическая функция распределения')

# Отображение графика
plt.show()
```
